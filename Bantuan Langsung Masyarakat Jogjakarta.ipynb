# Deskripsi

_Dataset ini berisi mengenai informasi mengenai BLM dan kemiskinan di DI Yogyakarta tahun 2007-2013. Dataset ini memiliki dimensi 282 baris dan 12 kolom_

__Penjelasan mengenai Variabel pada Dataset ini:__

1. kode_provinsi: Kode Provinsi

2. nama_provinsi: Nama Provinsi

3. kode_kabkota: Kode Kabupaten/Kota

4. nama_kabkota: Nama Kabupaten Kota

5. kecamatan: Nama Kecamatan

6. tahun: Tahun

7. jumlah_penduduk: Jumlah Penduduk di Kecamatan tersebut (berdasarkan PODES 2011)

8. jumlah_penduduk_miskin: Jumlah Penduduk Miskin di Kecamatan tersebut (berdasarkan PPLS 2011)

9. kategori: Kategori dari Kecamatan tersebut (Miskin/Sedang/Tidak Miskin)

10. blm_apbn: BLM (Bantuan Langsung Masyarakat) dari APBN (dalam Juta Rupiah)

11. blm_apbd: BLM (Bantuan Langsung Masyarakat) dari APBD (dalam Juta Rupiah)

12. program: Jenis PNPM

(https://data.go.id/dataset/data-bantuan-langsung-masyarakat-dan-kemiskinan-di-yogyakarta)

__Jenis tipe data__

- Kode Provinsi : interger

- Kecamatan : string

- Tahun : interger

- Program : String

- Jumlah Penduduk : interger

- Jumlah Penduduk Miskin : interger

- Kategori dari setiap kecamatan : string

- BLM dari APBN : Precision

__Penjelasan__

Bantuan Langsung Masyarakat dan Kemiskinan merupakan sebuah program dari Kementerian PPN atau Bappenas untuk pengentasan kemiskinan. Salah satunya diadakan di Yogyakarta. Rentang tahun pengumpulan data yaitu dari tahun 2007-2013. Kabupaten dan Kota yang berada di Jogjakarta antara lain, Kabupaten Bantul, Kabupaten Gunung Kidul, Kabupaten Kulon Progo, Kabupaten Sleman, dan Kota Yogyakarta. Data jumlah penduduk diambil berdasarkan data Potensi Desa (PODES). Sedangkan data jumlah penduduk miskin didasarkan pada data Program Perlindungan Sosial (PPLS). Baik PODES maupun PPLS merupakan program kerja Badan Pusat Statistik.

Diketahui dari grafik bahwa terdapat penigkatan baik jumlah penduduk maupun jumlah penduduk miskin dari tahun 2007 ke tahun 2008 kemungkinan disebakan oleh pencatatan baru (update data) yang dilakukan oleh BPS pada data PODES dan PPLS karena pada data BPS yang lain mengenai jumlah penduduk miskin menunjukkan tren membaik dari tahun 2007 ke 2010. Jumlah penduduk miskin yang diperoleh dari PPLS sangat berbeda dengan jumlah kemiskinan yang ditunjukkan oleh data BPS lainnya, misalnya tahun 2010 ke tahun 2011, data PPLS mencatat bahwa jumlah penduduk miskin yang ada di Yogyakarta tahun 2010 yaitu sebesar 332.835 jiwa dan tahun 2011 yaitu sebesar 315.579 jiwa. Sedangkan untuk data BPS lainnya mengenai tingkat kemiskinan di Yogyakarta menunjukkan bahwa tahun 2010 jumlah penduduk miskin yaitu sebesar 540.400 jiwa dan tahun 2011 yaitu sebesar 564.300 jiwa. Data PPLS menunjukkan adanya penurunan jumlah penduduk miskin sedangkan data BPS lainnya menunjukkan bahwa terdapat peningkatan jumlah penduduk miskin di Yogyakarta. Peningkatan jumlah penduduk miskin ini menurut BPS (2011) dikarenakan adanya laju inflasi sebesar 7,45% dalam periode Februari 2010 hingga Februari 2011.

Dari data diektahui bahwa jumlah penduduk miskin terbesar di Yogyakarta yaitu sebesar 13.848 jiwa yang berada di Kecamatan Gedang Sari, Kabupaten Gunung Kidul. Sedangkan jumlah penduduk miskin terkecil yaitu sebanyak 676 jiwa yang berada pada kecamatan Pakualaman, kota Yogyakarta. Jumlah Bantuan Langsung Masyarakat dan Kemiskinan Jogjakarta dari APBN yang terbesar adalah Rp 2850.0 juta di kecamatan Kalibawang, Kabupaten Kulon Progo tahun 2012. Jumlah Bantuan Langsung Masyarakat dan Kemiskinan Jogjakarta dari APBN yang terkecil adalah Rp 50.0 juta Kecamatan Ngampilan, Kota Yogyakarta tahun 2011. Jumlah Bantuan Langsung Masyarakat dan Kemiskinan Jogjakarta dari APBD yang terbesar adalah Rp 1400.0 juta salah satunya di Kecamatan Dlingo Kabupaten Bantul tahun 2008. Jumlah Bantuan Langsung Masyarakat dan Kemiskinan Yogyakarta dari APBD yang terkecil adalah Rp 7.5 juta di Kecamatan Ngampilan, Kota Yogyakarta di tahun 2013. Yang paling menarik adalah tercatatnya sejumlah penduduk dalam program PNPM Pasca Bencana namun tidak terdapat BLM dari APBN dan APBD yang dialokasikan untuk PNPM Pasca Bencana.
