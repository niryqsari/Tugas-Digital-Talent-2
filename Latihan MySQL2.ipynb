```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Report the account representative for each customer.
```

```python
run("""
SELECT 
    CONCAT(firstName,' ',lastName) AS fullName,
    customerName
FROM
    customers t1
        INNER JOIN
    employees t2 ON t1.salesRepEmployeenumber = t2.employeeNumber
ORDER BY fullName
""")
```

```python
#2. Report total payments for Atelier graphique.
```

```python
run("""
SELECT 
    customerNumber,
    customerName, 
    SUM(amount) AS TotalPembayaran
FROM
    payments p
        INNER JOIN
    customers c USING (customerNumber)
WHERE
    c.customerName = 'Atelier graphique'
""")
```

```python
#3. Report the total payments by date
```

```python
run("""
SELECT 
    paymentDate,
    SUM(amount) AS TotalPembayaran
FROM
    payments
GROUP BY paymentDate
""")
```

```python
#4. Report the products that have not been sold.
```

```python
run("""
SELECT 
    productName, 
    productLine, 
    productCode
FROM
    products
        LEFT JOIN
    orderdetails USING (productCode)
WHERE
    orderdetails.productCode IS NULL
""")
```

```python
#5. List the amount paid by each customer.
```

```python
run("""
SELECT 
    customerName, 
    customerNumber,
    amount 
FROM
    payments t1
        INNER JOIN
    customers t2 USING (customerNumber)
GROUP BY customerName
""")
```

```python
#6. How many orders have been placed by Herkku Gifts?
```

```python
run("""
SELECT 
    customerNumber,
    customerName, 
    COUNT(orderNumber) AS JumlahOrder
FROM
    orders o
        INNER JOIN
    customers c USING (customerNumber)
WHERE
    c.customerName = 'Herkku Gifts'
""")
```

```python
#7. Who are the employees in Boston?
```

```python
run("""
SELECT 
    officeCode,
    CONCAT(firstName,' ',lastName) AS fullName,
    city 
FROM
    employees e
        INNER JOIN
    offices o USING (officeCode)
WHERE
    o.city = 'Boston'
""")
```

8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT 
    customerName,
    SUM(amount) AS TotalPembayaran 
FROM
    payments p
        INNER JOIN
    customers c USING (customerNumber)
GROUP BY
    customerNumber
HAVING
    TotalPembayaran > 100000
ORDER BY 
    TotalPembayaran DESC
""")
```

```python
#9. List the value of 'On Hold' orders.
```

```python
run("""
SELECT 
    productCode,
    orderNumber,
    status,
    SUM(quantityOrdered * priceEach) AS value
FROM
    orderdetails od
        INNER JOIN
    orders o USING (orderNumber)
WHERE 
    o.status = 'On Hold'
GROUP BY productCode
ORDER BY orderNumber
""")
```

```python
#10. Report the number of orders 'On Hold' for each customer.
```

```python
run("""
SELECT 
    customerNumber,
    customerName,
    COUNT(orderNumber) AS TotalOrder,
    status 
FROM
    customers c
        INNER JOIN
    orders o USING (customerNumber)
WHERE 
    o.status = 'On Hold'
GROUP BY orderNumber
""")
```

__Many to Many Relationship__
1. List products sold by order date.
2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
4. Are there any products that appear on all orders?
5. List the names of products sold at less than 80% of the MSRP.
6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
7. List the products ordered on a Monday.
8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
#1. List products sold by order date.
```

```python
run("""
SELECT 
    orderDate,
    productCode,
    productName
FROM
    products p
        INNER JOIN
    orderdetails od USING (productCode)
        INNER JOIN
    orders o USING (orderNumber)
ORDER BY orderDate
""")
```

```python
#2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
```

```python
run("""
SELECT 
    productCode,
    productName,
    orderDate
FROM
    products p
        INNER JOIN
    orderdetails od USING (productCode)
        INNER JOIN
    orders o USING (orderNumber)
WHERE
    p.productName = '1940 Ford Pickup Truck'
ORDER BY orderDate DESC
""")
```

3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT 
    customerName,
    orderNumber,
    amount
FROM
    orders o
        INNER JOIN
    customers c USING (customerNumber)
        INNER JOIN
    payments p USING (customerNumber)
WHERE
    amount > 25000
ORDER BY amount DESC
""")
```

4. Are there any products that appear on all orders?

```python
run("""
SELECT 
    productName,
    COUNT(orderNumber) AS JumlahOrder
FROM
    orderdetails od
        INNER JOIN
    products p USING(productCode)
        INNER JOIN
    orders o USING(orderNumber)  
GROUP BY productName      
ORDER BY JumlahOrder
""")
```

5. List the names of products sold at less than 80% of the MSRP

```python
run("""
SELECT 
    productName, 
    priceEach, 
    MSRP
FROM 
    products p
        INNER JOIN
    orderdetails od USING (productCode)
WHERE 
    priceEach < (MSRP * .8)
ORDER BY priceEach DESC
""")
```

6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT 
    productName, 
    priceEach, 
    buyPrice
FROM 
    products p
        INNER JOIN
    orderdetails od USING (productCode)
WHERE 
    priceEach >= (2*buyPrice)
ORDER BY priceEach DESC
""")
```

7. List the products ordered on a Monday.

```python
run("""
SELECT 
    orderDate,
    productCode,
    productName, 
    orderNumber
FROM 
    products p
        INNER JOIN
    orderdetails od USING (productCode)
        INNER JOIN
    orders o USING (orderNumber)
WHERE 
    DAYNAME(orderDate) = 'Monday'
ORDER BY orderDate
""")
```

8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    productName,
    productCode,
    status,
    SUM(quantityOrdered) AS TotalonHold
FROM
    products p
        INNER JOIN
    orderdetails od USING (productCode)
        INNER JOIN
    orders o USING (orderNumber)
WHERE 
    o.status = 'On Hold'
GROUP BY productName
ORDER BY TotalonHold
""")
```
