```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SELECT * FROM offices ORDER BY country,state,city")
```

```python
run ("SELECT COUNT(employeeNumber) AS TotalKaryawan FROM employees")
```

```python
run("SELECT SUM(amount) AS TotalBayar FROM payments")
```

```python
run("""
SELECT 
    SUM(amount) AS TotalBayar 
FROM payments
WHERE paymentDate='2004-10-28'

""")
```

```python
run("""
SELECT 
    paymentDate,
    amount
FROM payments
WHERE amount>'100000'
ORDER BY paymentDate
""")
```

```python
run("""
SELECT 
    productName,
    productLine
FROM products
ORDER BY productLine ASC
""")
```

```python
run("""
SELECT 
    productLine,
    COUNT(*) AS JumlahProduk
FROM products
GROUP BY productLine
ORDER BY productLine
""")
```

```python
run("""
SELECT 
    MIN(amount) AS MinimumPayments
FROM payments
""")
```

```python
run("""
SELECT 
    paymentDate,
    amount
FROM 
    payments
WHERE 
    amount > 2 * (select AVG(amount) from payments)
ORDER BY paymentDate
""")
```

```python
run("""
SELECT 
    AVG(((MSRP-buyPrice)/buyPrice)*100) AS AverageMarkupPercn
FROM 
    products
""")
```

```python
run("""
SELECT 
    COUNT(DISTINCT productCode) AS TotalDistinctProducts
FROM 
    orderdetails 
""")
```

```python
run ("SELECT customerName, country FROM customers WHERE salesRepEmployeeNumber IS NULL")
```

```python
run("""
SELECT 
    CONCAT (firstName," ",lastName) AS FullName,
    jobTitle
FROM 
    employees
WHERE 
    jobTitle LIKE '%Manager%' OR jobTitle LIKE '%VP%' 
""")
```

```python
run("""
SELECT 
    orderNumber,
    SUM(quantityOrdered*priceEach) AS TotalPrice
FROM 
    orderdetails
WHERE 
    (quantityOrdered*priceEach)>5000
GROUP BY orderNumber
ORDER BY TotalPrice
""")
```
